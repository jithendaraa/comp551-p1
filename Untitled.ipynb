```python
import os
import pandas as pd
import matplotlib.pyplot as plt

rootdir = 'Adult dataset'
dirs = os.listdir(rootdir)
print(dirs)
```

```python
adult_names = ["age",
"workclass",
"fnlwgt",
"education",
"education-num",
"marital-status",
"occupation",
"relationship",
"race",
"sex",
"capital-gain",
"capital-loss",
"hours-per-week",
"native-country"]
```

```python
adult_data = pd.read_csv(os.path.join(rootdir, 'adult.data'), names=adult_names, index_col=False)
adult_data.head()
```

```python
for col_name in adult_names:
    print(col_name, adult_data[col_name].isna().values.any())
```

```python
df = adult_data.drop_duplicates()
len(df)
```

```python
col_names_with_nan = []
for col_name in adult_names:
    if ' ?' in adult_data[col_name].unique().tolist():
        col_names_with_nan.append(col_name)
        print(col_name)
        print(adult_data[col_name].unique().tolist())
        print()
```

## Remove ' ?' in workclass, occupation, native-country

```python
df = adult_data
print(len(df))
for col_name in col_names_with_nan:
    df = df[getattr(df, col_name) != ' ?']
```

```python


n = len(df.select_dtypes(include=['object']).columns.tolist())
ncols = 4
plt.figure(figsize=(16 * n // ncols, 4 * ncols))
for idx, col in enumerate(df.select_dtypes(include=['object']).columns.tolist()):
    plt.subplot(n // ncols, ncols, idx  + 1)
    plt.bar( df[col].value_counts().index, df[col].value_counts().values)
    plt.title(col)
    plt.grid()
    # if idx >= 3: break
plt.tight_layout()
plt.show()
```
